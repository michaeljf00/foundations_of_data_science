Understanding the code line above

* **smf** calls the package StatsModels
* **ols** tells Python we are using an Ordinary Least Square (OLS) regression (a type linear regression)
* **formula=** used to write the dependent and all the independent variable(s)
* **first variable inside the parenthesis before "~"/dependent variable/outcome**: The first variable is our only dependent variable. This is our outcome, the variable that determines which type of regression to use;
* **~ inside parenthesis**: Marks the border between the outcome (dependent variable) to the left, and the independent variables to the right;
* **independent variables** : All other variables after the "~", inside parenthesis;
* **+ sign inside parenthesis**: the + sign is used to separate different independent variables inside the same model
* **C()** marks a categorical variable
* **,data=** This marks the name of the data frame.
* **.fit()** tells Python we want to fit our function ("run the function")

# **Predictive Model**


```python
from sklearn.linear_model import LinearRegression # for running linear regression model
from sklearn.model_selection import train_test_split # for splitting the data into train and test
```

**Preparing data for predictive modeling **

Since a predictive model will be evaluated only in the future when we are able to compare our predictions for the future with the actual value relaized in future, we can't evaluate a predictive model w/o waiting for the future.

Thus, we break the data into two parts train (to fit the model) and test (to test the model's prediction accuracy). We assume that the test data is the future data.

Thus, we follow the following steps in predictive models:

1. Prepare the data
   * Dummy code (one-hot encode) the categorical variables
   * Separate the independent variables (called as predictors in predictive modeling) and the dependent/ outcome variable (called as target)

2. Break the data into train and test sets

3. Fit the Linear Regression model on the train data
   
4. Evaluate the model
   * Make predictions on the train data, and evaluate the model on train data
   * Make predictions for the test data, and evaluate the model on test data



## Data Preparatation

We already checked for NA values and thus need not repeat here.

Dummy code the categorical variables using "pd.get_dummies"

```python
df_dummies = pd.get_dummies(df, drop_first = True) # "drop_first = True" drops one of the categories
```

Create X, a set of all predictors and y, our target variable.

```python
X = df_dummies.drop(columns = ["price"]) # always follow the standard practice of calling it capital X

y = df_dummies["price"] # always follow the standard practice of calling it small y
```

**Train-test split**

Keep all X capitalized, and y small.

**test_size**  sets the % of the data that goes into test set.

**random_state** ensures replicability of the splitting. If we don't fix it to a number, everytime we do train-test split, we may get different rows in the train/ test datasets. For now, just set it to what i do so that you may replicate my results.

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2,  random_state = 1) #
```

## Model Fitting

Model Initialization

Model Fitting - We pass two arguments here: train_X (set of predictors) and train_y(target)

```python
pred_model = LinearRegression() # initialize the model with the name pred_model

pred_model = pred_model.fit(train_X, train_y)

# we will use this model "pred_model" to make predictions
```

## Making Predictions

On train data

On test data

```python
train_predictions = pred_model.predict(train_X)

test_predictions = pred_model.predict(test_X)
```

## Calculating Errors  

On train data

On test data

```python
train_errors = train_y - train_predictions

test_errors = test_y - test_predictions
```

```python
#### Training ####
train_ME = round(np.mean(train_errors), 2)
train_MAE = round(np.mean(abs(train_errors)), 2)

train_MPE =   round(np.mean(train_errors / train_y), 5)  * 100
train_MAPE =  round(np.mean(abs(train_errors) /  train_y), 5) * 100

train_RMSE = round(np.sqrt(np.mean(train_errors*train_errors)), 2)

train_metrics =  [train_ME, train_MPE, train_MAE, train_MAPE, train_RMSE]

#### Test ####

test_ME = round(np.mean(test_errors), 2)
test_MAE = round(np.mean(abs(test_errors)), 2)

test_MPE = round(np.mean(test_errors / test_y), 5) * 100
test_MAPE = round(np.mean(abs(test_errors) / test_y), 5) * 100

test_RMSE = round(np.sqrt(np.mean(test_errors*test_errors)),2)

test_metrics =  [test_ME, test_MPE,  test_MAE,test_MAPE, test_RMSE]
```

```python
predictive_model_performance = pd.DataFrame({"Error Metric": ["ME", "MPE",  "MAE","MAPE", "RMSE"], "Train": train_metrics, "Test": test_metrics})
```

```python
predictive_model_performance
```

You must be wondering why are we evaluating the model on both Train and Test data. By comparing the model performance in the two datasets, we can measure the degree of overfitting in our model. I will explain the concept next week.
